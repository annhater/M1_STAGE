```python
#IMPORT
import os
import numpy as np
import pandas as pd
from arcplot import ArcPlot, show_arc_plot, save_arc_plot_as
import matplotlib.pyplot as plt
```

```python
#SETUP
rootdir = '/content/drive/MyDrive/M1_STAGE/Data/simulations_1HSI'
os.chdir(rootdir)
```

```python
#MAIN

# 1. Read data

# 1.1. Open frequency tables, read as df
contact_maps_paths = []

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        if file.startswith("contact_"):
	        contact_maps_paths.append(os.path.join(subdir, file))


for sim in contact_maps_paths:
    sim_name = sim.replace("contact_", "").replace(".csv", "")
    df_sim = pd.read_csv(sim)
    title = f'Interactions intra-atomiques pour {sim_name}'
    src_path  = r"/content/drive/MyDrive/M1_STAGE/Manips/"
    p = Path(src_path).parent.joinpath(f"Manips/Figures/{sim_name}_arcgraph.png")
    show_arc_plot(df_sim,
                title=title,
                node_size=120,
                width=5,
                height=5)
    save_arc_plot_as(df_sim,
                title=title,
                file_name=p,
                node_size=120,
                resolution=100,
                width=19.2,
                height=10.8)
```
